## keras OCR

```python
import matplotlib.pyplot as plt
```

```python
import keras_ocr
import os

# Initialize the OCR recognizer and pipeline
recognizer = keras_ocr.recognition.Recognizer()
pipeline = keras_ocr.pipeline.Pipeline(recognizer=recognizer)

# Path to weights and image
weights_path = 'C:\\Users\\sarth\\.keras-ocr\\craft_mlt_25k.h5'
image_path = r'data\id_image_preprocessed.png'

# Check if weights file exists
if not os.path.exists(weights_path):
    raise FileNotFoundError(f"Weights file not found at {weights_path}")

# Check if image file exists
if not os.path.exists(image_path):
    raise FileNotFoundError(f"Image file not found at {image_path}")

# Load weights for the recognizer
recognizer.model.load_weights(weights_path)

# Read images from folder path to image object
images = [keras_ocr.tools.read(image_path)]

# Perform OCR on the images
prediction_groups = pipeline.recognize(images)

# Display the OCR results
for predictions in prediction_groups:
    for text, box in predictions:
        print(f"Text: {text}, Box: {box}")
```

```python
# generate text predictions from the images
prediction_groups = pipeline.recognize(images)

# plot the text predictions
fig, axs = plt.subplots(nrows=len(images), figsize=(10, 20))
for ax, image, predictions in zip(axs, images, prediction_groups):
    keras_ocr.tools.drawAnnotations(image=image, 
                                    predictions=predictions, 
                                    ax=ax)
```

## paddle OCR
Need to create a seperate environment for Paddle OCR because its libraries dependencies does not match with other libraries.

```python
from paddleocr import PaddleOCR, draw_ocr

def show_ocr(ocr, img_path, font):

    result = ocr.ocr(img_path)
    # save_path = os.path.join(out_path, img_path.split('/')[-1] + 'output')

    image = cv2.imread(img_path)

    boxes = [line[0] for line in result[0]]
    txts = [line[1][0] for line in result[0]]
    scores = [line[1][1] for line in result[0]]

    im_show = draw_ocr(image, boxes, txts, scores, font_path=font)

    # cv2.imwrite(save_path, im_show)

    img = cv2.cvtColor(im_show, cv2.COLOR_BGR2RGB)
    plt.imshow(img)	
    plt.axis('off')
    # return result
```

```python
ocr = PaddleOCR(use_angle_cls=True, use_gpu=False)
img_path = "data\id_image_preprocessed.png"
font = "model\simfang.ttf"
show_ocr(ocr, img_path,font)
```
